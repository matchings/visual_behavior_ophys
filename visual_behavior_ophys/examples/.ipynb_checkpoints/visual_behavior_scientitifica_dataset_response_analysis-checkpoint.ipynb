## ResponseAnalysis object has several useful summary dataframes 

### response_df has a row for every trial for every cell

```python
ra.response_df.head()
```

it includes a snippet of the trace for each trial for a window defined by ra.response_window around the time of the stimulus change, as well as running speed over the same duration, p_value of response, and behavioral metadata

### mean response dataframe has a row for every unique combination of stimulus (change_code) and behavioral response (HIT, MISS, FA, CR) for every cell. 

```python
ra.mean_response_df.head()
```

For each condition, it includes the mean trace, the mean within the response window (500ms after stim onset), and significance of the response (as a p_value comparing response window to baseline (500ms before stim onset) and as the # of standard deviations for which the response exceeds the baseline values)

### cell summary dataframe has a row for every cell

```python
ra.cell_summary_df.head()
```

it includes summary statistics, including preferred stimulus, p_value and sd_over_baseline for preferred condition, reliability of response, and number of responsive conditions

#### you can generate a summary dataframe with different thresholds for responsiveness if you want (default is that p_value for pref condition has to be  0.005 and sd_over_baseline has to be > 3)

```python
lenient_sdf = ra.get_cell_summary_df(p_val_thresh = 0.05, sd_over_baseline_thresh = 2)
```
